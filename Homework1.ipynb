```python
import pandas as pd
```

```python
df = pd.read_csv('archive.zip', index_col=0)
```

```python
df.head()
```

```python
numerical_columns = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
numerical_descriptions = df[numerical_columns].describe()
print(numerical_descriptions)
```

```python
categorical_columns = ['cut', 'color', 'clarity']
categorical_descriptions = df[categorical_columns].describe()
print(categorical_descriptions)
```

```python
unique_cut_values = df['cut'].unique()
unique_color_values = df['color'].unique()
unique_clarity_values = df['clarity'].unique()

print("Unique cut values:", unique_cut_values)
print("Unique color values:", unique_color_values)
print("Unique clarity values:", unique_clarity_values)
```
